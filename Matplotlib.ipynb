```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
plt.plot(t, f, color="green")
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

Задание 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title("График y2")
ax3.plot(x, y3)
ax3.set_title("График y3")
ax4.plot(x, y4)
ax4.set_title("График y4")
```

```python
plt.style.use('fivethirtyeight')
path_csv = "/content/drive/MyDrive/Colab Notebooks/creditcard.csv"
data = pd.read_csv(path_csv)
```

```python
Class_counts = pd.DataFrame(data.Class.value_counts())
Class_counts.plot(kind="bar", xlabel="Значения Class", ylabel="кол-во наблюдений")
Class_counts.plot(kind="bar", logy=True, xlabel="Значения Class", ylabel="кол-во наблюдений")
```

```python
V1_1 = data[data['Class']==1]['V1']
V1_0 = data[data['Class']==0]['V1']
plt.hist([V1_0, V1_1], density=True, bins=20, alpha=0.5, label=['Class 1', 'Class 0'], color=['red', 'gray'])
plt.legend()
plt.xlabel("Значения V1", fontsize=14)
plt.ylabel("Плотность распределения", fontsize=14)
```
